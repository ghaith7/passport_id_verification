```python
from easyocr import Reader
import cv2
import matplotlib.pyplot as plt 
```

```python
import os
os.environ['KMP_DUPLICATE_LIB_OK']='True'
```

```python
image = "testing.jpg"
langs = ["en","ar"]
```

```python
im = cv2.imread(image)
plt.imshow(im)
plt.show()
```

```python
reader = Reader(langs, gpu=True)
results = reader.readtext(image)
for (bbox, text, prob) in results:
    print(text)
```
